### Isolation Forest  

Classical anomaly detection requires the measurement of all data points to establish a multivariate Gaussian distribution, which can then be used to identify regions of low density. In other words, large amounts of data have to be processed in-order to establish a bar for what normal means. The methods then use the deviation from normality as an outlier detection metric. Although this appears to be a natural way to solve the outlier problem, it requires one to process huge amounts of data, which for some datasets is computationally not feasible.

Isolation Forest offers an ingenious, fast and light weight alternative for identifying outliers without first going through the laborious task of establishing what normal means, and often outperforms more technical methods such as single-class support vector machines (SVMs). The figure below demonstrates the idea behind this method. For more details see the original paper by <font color='dodgerblue'>Liu et al. IEEE 2008</font>. 

<img src="pics/outlier.png" width="1000">   

***Method:*** The two panels on the left show a blue inlier (normal point) and a red outlier. The Isolation Forest algorithm draws a random line through the data space by 1) randomly selecting an axis (x- or y-axis), and then 2) randomly selecting a position on the axis to intersect (split point). Populating the space with these lines is referred to as "growing a tree". Each time a new line is introduced, it partitions the data set into subsets, and at some point, each data instance is isolated into its own box. Here is the logic. It will naturally take fewer random lines to partition an outlier than an inlier. Therefore, the number of lines necessary to isolate a point can be used as an outlier metric. Specifically, the length of the tree needed to isolate a point is indirectly proportional to the outlier score. If one is familiar with graph theory, this process is equivalent to forming a binary search tree. At some point each data instance will appear on a "leaf", and the outliers will all be found at the root.  

***A caveat:*** Since growing a tree is predicated on a random two step binary decision mechanism, it is possible, but not probable, that we isolate a inlier before an outlier simply by chance. Isolation Forest therefore grows many trees, called a "Forest", and then takes the aggregate score for each data point. The score can be seen to converge in the right hand panel for two different IRIS spectra.  

***Advantage:*** Since the outliers are all captured early in the tree growing process, one does not have to grow the tree all the way to its conclusion, and can simply truncate. Meaning, even with a large dataset, we just have to "skim the top".  

The two images on the left are taken from <font color='dodgerblue'>Liu et al. 2008</font>.

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope
np.random.seed(1)
```

In this notebook we will show an example of a simple dataset where Isolation Forest outperforms another method. We will then try to recreate the plot appearing above in the right hand panel by applying the Isolation Forest algorithm to actual IRIS data.

```python
# Create data for circle using sklearn datasets class
n_samples = 500
X, y = datasets.make_circles(n_samples=n_samples, factor=.05, noise=.07)
# This class creates an inner and out circle, remove the inner circle
X = X[ np.squeeze( np.argwhere( y==0 )), : ] 
```

```python
# Add a data point in the center, this will be our outlier
outlier = np.array([[0,0]])
X = np.concatenate( (X, outlier), axis=0 )
```

```python
# Plot our synthetic dataset
fig, ax = plt.subplots(figsize=(8,8))
plt.scatter(X[:,0], X[:,1], s=25, c='k')
plt.xticks([])
plt.yticks([])
plt.show()
```

Clearly the outlier here is the single point at the circles center, but this is also the average of all the points. Although this seems like a tricky unnatural dataset, it is not uncommon for outliers to surround and shield themselves in a chamber like this, masquerading in between the folds of data as the most normal, inconspicuous of all members. Lets see if a robust covariance approach can identify the outlier.

```python
# Covariance method
cov = EllipticEnvelope(random_state=0).fit( X )
scores  = cov.predict( X )
```

```python
fig, ax = plt.subplots(figsize=(8,8))
# Create inlier and outlier masks
inlier = scores == 1
outlier = scores == -1
plt.scatter(X[inlier, 0], X[inlier, 1], c='dodgerblue', s = 25, label='inlier')
plt.scatter(X[outlier, 0], X[outlier, 1], c='coral', s = 25, label='outlier')
plt.xticks([])
plt.yticks([])
plt.legend(loc="best")
lgnd = plt.legend(loc="upper left", numpoints=1, fontsize=15)
plt.show()
```

This method fails terribly, and cannot identify the obvious outlier. Next up we try the Isolation Forest.

```python
# Apply outlier Isolation Forest
clf = IsolationForest(random_state=0,  n_estimators=200).fit( X )
scores  = clf.predict( X ) # scores  = clf.decision_function( X )
```

```python
fig, ax = plt.subplots(figsize=(8,8))
# Create inlier and outlier masks
inlier = scores == 1
outlier = scores == -1
plt.scatter(X[inlier, 0], X[inlier, 1], c='dodgerblue', s = 25, label='inlier')
plt.scatter(X[outlier, 0], X[outlier, 1], c='coral', s = 25, label='outlier')
plt.xticks([])
plt.yticks([])
plt.legend(loc="best")
lgnd = plt.legend(loc="upper left", numpoints=1, fontsize=15)
plt.show()
```

Success! The central point is flagged as an outlier, and furthermore has the highest score by far when using the decision_function method.

### Lets import the two spectral datasets
1. Mg II spectra collected from a solar flare (FL) <font color='coral'>FL_20140910_112825_3860259453</font>
2. Mg II spectra from the quiet Sun (QS) <font color='dodgerblue'>QS_20151010_233413_3620005935</font>    

The spectra for the flare dataset is stored in a numpy array called <font color='coral'>fl_spectra</font>. Each row is a different spectra of length 240. Similarly, the quiet Sun dataset is stored in a numpy array called <font color='dodgerblue'>qs_spectra</font>. Each spectrum can be described in terms of several features found in the variables <font color='coral'>fl_features</font> and <font color='dodgerblue'>qs_features</font>.

```python
# Load quiet Sun dataset
with open('data/qs_spectra.p', 'rb') as f: qs_spectra = pickle.load(f)
with open('data/qs_features.p', 'rb') as f: qs_features = pickle.load(f)
# Load flare dataset
with open('data/fl_spectra.p', 'rb') as f: fl_spectra = pickle.load(f)
with open('data/fl_features.p', 'rb') as f: fl_features = pickle.load(f) 
```

## ! Exercise solutions

```python
'''
Q1) Join a single flare (index 91000) spectrum to the end of the quiet Sun dataset
'''
outlier = fl_spectra[91000].reshape(1, fl_spectra.shape[1])
X = np.concatenate( (qs_spectra, outlier), axis=0 )
```

```python
'''
Q2) Show that the flare spectrum, which is the last index, is assigned a - 1 by the algorithm, meaning outlier 
    (+1 is for inlier)
'''
clf = IsolationForest(random_state=0, n_estimators=200).fit( X )
scores  = clf.predict( X ) # scores  = clf.decision_function( X )
print(scores)
```

```python
'''
Q3) Evaluate how the score converges for the flare spectrum a normal quiet Sun inlier as we change the number of trees

To reconcile the results to the original paper, transform the cores with scores = (2 ** -scores) -0.5
Values < 0.5 are safely inliers
Values close to 1 are outliers
'''
# Select the most normal spectrum index to follow
inlier_ind = np.argwhere(abs(scores - np.mean(scores)) == min( abs(scores - np.mean(scores)) ))[0][0] #1850
inlier = X[inlier_ind]

# Select the flare outlier spectrum index to follow
outlier_ind = len(X) - 1
outlier = X[outlier_ind]


# Track the change in score for these two indices while increasing the number of trees (base estimators) in our forest
inlier_score  = []
outlier_score = []
for n_trees in np.linspace(1,100,100):
    clf = IsolationForest(random_state=0, max_features=20, n_jobs=-1, max_samples=1000, n_estimators=int(n_trees)).fit( X )
    scores  = clf.decision_function( X )
    scores = (2 ** -scores) -0.5
    inlier_score.append( scores[inlier_ind] )
    outlier_score.append( scores[outlier_ind] )
```

```python
'''
Q4) Plot the results
'''
fig, ax = plt.subplots( figsize=(10,6) )
xax = np.linspace(1,100,100)
plt.plot(xax, inlier_score, c='dodgerblue', label='inlier score')
plt.plot(xax, outlier_score, c='coral', label='outlier score')
plt.legend(loc="best")
plt.xlim(0,100)
plt.ylim(0.35,0.85)
plt.show()
```
